```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
import extract_lat as el
import glob
import os
```

```python
# Data/moire or Data/atomic
os.chdir('Data/atomic')
df = glob.glob('*.txt')
df
```

```python
a={}
x={}
y={}
z={}
peak={}
peakz={}
i=2
```

```python
a[df[i]],x[df[i]],y[df[i]],z[df[i]],peak[df[i]]=el.lat_con(df[i],pxdist=5,thrs=0.002)
```

```python
peakz[df[i]]=np.ones_like(peak[df[i]][:,0])
fig = plt.figure()
plt.gca().set_aspect('equal')
ax = plt.axes(projection="3d")
ax.plot_surface(y[df[i]],x[df[i]],z[df[i]],cmap=plt.cm.gray,linewidth=0, antialiased=False, rcount=200, ccount=200 )
ax.scatter3D(peak[df[i]][:, 0], peak[df[i]][:, 1],peakz[df[i]]*3,'.',s=2,color='red')
ax.view_init(90, 0)
#plt.xlim(0,5)
#plt.ylim(0,5)
#plt.axis('off')
```
