```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

Lattice vectors and origin from data. user defined total number of repeatition along each vector

```python
# values from fft of  F:\Data\4K\March2018\18c30\18c23T0014.SM4
n1=60
n2=60
v1_ini = [-0.462691, -0.209270]
v2_ini = [-0.024471,-0.522391]
ori=[1.1941,0.9856]
# values from lattice of  F:\Data\4K\March2018\18c30\18c23T0014.SM4
# v1_ini = [-0.441022, -0.212932]
# v2_ini = [-0.0141311,-0.525647]
# ori=[1.1941,0.9856]
```

```python
rot=0
theta = np.radians(rot)
c, s = np.cos(theta), np.sin(theta)
R = np.array(((c, -s), (s, c)))
```

Generating x and y coordinates from index of each point

```python
v1=R.dot(v1_ini)
v2=R.dot(v2_ini)
```

```python
pt_index = np.array(np.meshgrid(np.arange(-n1,n1),np.arange(-n2,n2)))
x= pt_index[0,:,:]* v1[0]+ pt_index[1,:,:]* v2[0]+ori[0]
y= pt_index[0,:,:]* v1[1]+ pt_index[1,:,:]* v2[1]+ori[1]
z=np.ones_like(x)
```

Masking x and y values to get lattice in chosen x,y range

```python
mask_x = np.logical_and(x > 0, x < 28)
mask_y = np.logical_and(y > 0, y < 28)
mask_z = ~np.logical_and(mask_x,mask_y)
z_masked=np.ma.masked_array(z, mask_z)
```

```python
# atomic resolution on substrate
data=np.loadtxt('Data/18c23T0014crop.txt')
data_z=data*1e9
data_len=1.87
data_x=np.linspace(0, data_len, num=data_z.shape[0])
data_y=data_x
X,Y=np.meshgrid(data_x,data_y)
```

```python
fig = plt.figure()
plt.gca().set_aspect('equal')
ax = plt.axes(projection="3d")
ax.plot_surface(X,Y,data_z,cmap=plt.cm.gray,linewidth=0, antialiased=False, rcount=200, ccount=200 )
ax.scatter3D(x,y,z_masked,'.',s=2,color='red')
ax.view_init(90, 0)
#plt.xlim(0,7)
#plt.ylim(0,7)
```

```python
# Atomic res on island 
# values from fft of  F:\Data\4K\March2018\18d02\18d02T0004.SM4
n1=100
n2=100
iv1_ini = [0.240420, 0.344074]
iv2_ini = [-0.189432,0.401031]
ori=[1.1941,0.9856]
# values from lattice of   F:\Data\4K\March2018\18d02\18d02T0004.SM4
# iv1_ini = [0.241786, 0.366734]
# iv2_ini = [-0.189115,0.405762]
# ori=[1.1941,0.9856]
```

```python
rot=0
theta = np.radians(rot)
c, s = np.cos(theta), np.sin(theta)
R = np.array(((c, -s), (s, c)))
iv1=R.dot(iv1_ini)
iv2=R.dot(iv2_ini)
pt_index = np.array(np.meshgrid(np.arange(-n1,n1),np.arange(-n2,n2)))
ix= pt_index[0,:,:]* iv1[0]+ pt_index[1,:,:]* iv2[0]+ori[0]
iy= pt_index[0,:,:]* iv1[1]+ pt_index[1,:,:]* iv2[1]+ori[1]
iz=np.ones_like(ix)
imask_x = np.logical_and(ix > 0, ix < 28)
imask_y = np.logical_and(iy > 0, iy < 28)
imask_z = ~np.logical_and(imask_x,imask_y)
iz_masked=np.ma.masked_array(iz, imask_z)
```
